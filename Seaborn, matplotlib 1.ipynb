```python
import seaborn as sns 
import matplotlib as plt 
import numpy as np
import pandas as pd 
```

```python
tips = pd.read_csv("data/tips.csv")
tips
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips )
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time', data = tips ) #hue for color 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', style = 'time', data = tips ) #style for diff shape 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', size = 'time', data = tips ) #size for dot size 
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", size="day", style="time", data=tips) #combine all 
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips) # if the value is continuous, the color will be in gradient 
```

```python
df = pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum()})
```

```python
sns.relplot(x="time", y="value", kind="line", data=df) #use 'kind' for line to create visual for time series 
#by default 'kind'= scatter 
```

```python
df = pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum(), 
    "region": "North", "division": "A"})
df = df.append(pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum(), 
    "region": "North", "division": "B"}))
df = df.append(pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum(), 
    "region": "North", "division": "C"}))
df = df.append(pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum(), 
    "region": "South", "division": "A"}))
df = df.append(pd.DataFrame({
    "time": np.arange(500),
    "value": np.random.randn(500).cumsum(), 
    "region": "South", "division": "B"}))

sns.relplot(
    x="time", y="value", kind="line", hue="region", 
    style="division", data=df)  #combine hue(color), and style(shape)
```

```python
df = pd.DataFrame({
    "time": np.arange(20),
    "value": np.random.randn(20).cumsum(), 
    "region": "North"})
df = df.append(pd.DataFrame({
    "time": np.arange(20),
    "value": np.random.randn(20).cumsum(), 
    "region": "South"}))
sns.relplot(x="time", y="value", kind="line", 
            style="region", markers=True, data=df)  #use markers for dot when plotting 
```

```python
sns.relplot(x="time", y="value", kind="line", style="region", 
            dashes=False, markers=True, data=df)  #and dashes for doted line
```
